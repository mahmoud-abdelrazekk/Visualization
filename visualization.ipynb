```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
movies = pd.read_csv("movies_dataset.csv")
movies.dropna(inplace=True)
penguins = sns.load_dataset("penguins")
penguins.dropna(inplace=True)
```

```python
yearly_revenue = movies.groupby('year')['revenue_million'].mean()
plt.plot(yearly_revenue.index, yearly_revenue.values, marker='o')
plt.title('Average Movie Revenue Per Year')
plt.xlabel('Year')
plt.ylabel('Average Revenue (in millions)')
plt.grid(True)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(movies['votes'], movies['rating'])
ax1.set_title('Rating vs. Votes')
ax1.set_xlabel('Votes')
ax1.set_ylabel('Rating')
ax1.grid(True)
ax2.scatter(movies['budget_million'], movies['revenue_million'], color='orange')
ax2.set_title('Budget vs. Revenue')
ax2.set_xlabel('Budget (in millions)')
ax2.set_ylabel('Revenue (in millions)')
ax2.grid(True)
```

```python
plt.hist(penguins['body_mass_g'], bins=20, edgecolor='black', color='skyblue')
plt.title('Distribution of Penguin Body Mass')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')
```

```python
yearly_counts = movies['year'].value_counts().sort_index()
plt.bar(yearly_counts.index, yearly_counts.values, color='green')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
```

```python
sns.boxplot(x='species', y='body_mass_g', data=penguins)
plt.title('Penguin Body Mass Across Species')
```

```python
sns.violinplot(x='species', y='flipper_length_mm', data=penguins)
plt.title('Flipper Length Across Penguin Species')
```

```python
sns.histplot(data=movies, x='rating', bins=10, kde=True)
plt.title('Distribution of Movie Ratings')
```

```python
sns.barplot(x='genre', y='revenue_million', data=movies, estimator=lambda x: sum(x)/len(x))
plt.title('Average Revenue Per Movie Genre')
```

```python
sns.countplot(x='genre', data=movies, order=movies['genre'].value_counts().index)
plt.title('Number of Movies by Genre')
```

```python
numeric_penguins = penguins.select_dtypes(include=['float64', 'int64'])
sns.pairplot(numeric_penguins)
```

```python
corr_cols = ['budget_million', 'revenue_million', 'rating', 'votes']
corr_matrix = movies[corr_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Movie Metrics')
```

```python
g = sns.FacetGrid(penguins, col="species")
g.map(sns.scatterplot, "bill_length_mm", "bill_depth_mm")
```

```python
sns.regplot(x='budget_million', y='revenue_million', data=movies)
plt.title('Regression Plot of Budget vs. Revenue')
```

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins, kind='reg')
```

```python
sns.set_theme(style="ticks", palette="pastel")
sns.boxplot(x='species', y='body_mass_g', data=penguins)
plt.title('Penguin Body Mass Across Species (Custom Style)')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
```

```python
plt.hist(movies['rating'], bins=15, color='purple', edgecolor='darkblue', alpha=0.7)
plt.title('Distribution of Movie Ratings (Customized)')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y')
```

```python
sns.regplot(x='budget_million', y='rating', data=movies)
plt.title('Does a Bigger Budget Lead to a Higher Rating?')
```

```python
sns.pairplot(penguins, hue='species')
plt.suptitle('Pairplot of Penguin Measurements by Species', y=1.02)
```
